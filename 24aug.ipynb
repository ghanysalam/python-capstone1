# Merge

```python
salaries = sql_df('''
select * from salaries
''')
salaries
```

```python
titles = sql_df('''
select * from titles
''')
titles
```

```python
employees = sql_df('''
select * from employees
''')
employees
```

```python
departments = sql_df('''
select * from departments
''')
departments
```

```python
dept_emp = sql_df('''
select * from dept_emp
''')
dept_emp
```

```python
emp_titles = pd.merge(employees, titles, on = 'emp_no')
emp_titles
```

```python
emp_titles = pd.merge(employees, titles, on = 'emp_no', how='right')
emp_titles
```

```python
emp_titles = pd.merge(employees, titles, on = 'emp_no', how='left')
emp_titles
```

```python
concat_emp = pd.concat([employees, titles])
concat_emp[concat_emp['emp_no']==10001]
```

```python
pd.concat([employees, titles], axis =1)
```

```python
emp_dept = pd.merge(employees, dept_emp, on='emp_no').merge(departments, on='dept_no')
emp_dept
```
